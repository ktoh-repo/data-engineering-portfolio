```
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .config("spark.jars.packages", "com.crealytics:spark-excel_2.12:0.13.5") \
    .getOrCreate()
```

```
df = spark.read.format("com.crealytics.spark.excel") \
    .option("inferSchema", "false") \
    .option("header", "true") \
    .option("dataAddress", "'Dirty 7'!A1") \
    .load("/FileStore/tables/6__Hospital_Data_with_Mixed_Numbers_and_Characters.xlsx")

display(df)
```

```
from pyspark.sql.functions import col, sum, when
```

```
null_count_df = df.select(
    [sum(when(col(c).isNull(), 1).otherwise(0)).alias(c) for c in df.columns]
)
null_count_df.show()
```

```
from pyspark.sql.functions import regexp_replace, col
```

```
def str_to_num(df, column_name):
    df= df.withColumn(column_name, regexp_replace(col(column_name), "[o/O]", "0"))
    df= df.withColumn(column_name, regexp_replace(col(column_name), "[iI]", "1"))
    df= df.withColumn(column_name, regexp_replace(col(column_name), "[sS]", "5"))
    df= df.withColumn(column_name, regexp_replace(col(column_name), "/D", ""))

    return df
```

```
df = str_to_num(df, "Beginning Balance ")
```

```
df = str_to_num(df, "Quantity Received ")
```

```
df = str_to_num(df, "Quantity Dispensed ")
```

```
df = str_to_num(df, "Losses and Adjustments")
```

```
df = str_to_num(df, "Ending Balance (Physical Count)")
```
