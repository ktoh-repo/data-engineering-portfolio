```python
import pandas as pd
```

```python
df = pd.read_excel("../3.-Badly-Structured-Sales-Data-3.xlsx", header=1, skipfooter=1)
```

```python
df.head()
```

```python
df.columns
```

```python
df.rename(columns={'Unnamed: 0': "Order ID", 'Segment': "Order Date",
                   "Consumer": "Consumer First Class", "Corporate": "Corporate First Class", "Home Office": "Home_Office First Class" ,
                   "Consumer.1": "Consumer Same Day", "Corporate.1": "Corporate Same Day", "Home Office.1": "Home_Office Same Day" ,
                   "Consumer.2": "Consumer Second Class", "Corporate.2": "Corporate Second Class", "Home Office.2": "Home_Office Second Class" ,
                   "Consumer.3": "Consumer Standard Class", "Corporate.3": "Corporate Standard Class", "Home Office.3": "Home_Office Standard Class" ,
                  }, inplace=True)
```

```python
df = df.melt(id_vars=["Order ID", "Order Date"], value_vars=["Consumer First Class", "Corporate First Class", "Home_Office First Class",
                                                       "Consumer Same Day", "Corporate Same Day", "Home_Office Same Day",
                                                       "Consumer Second Class", "Corporate Second Class", "Home_Office Second Class" ,
                                                       "Consumer Standard Class","Corporate Standard Class", "Home_Office Standard Class"
                                                       ], var_name="Segment and Ship Mode", value_name="Sales")
```

```python
df = df.drop(index=0)
```

```python
df = df.dropna(subset=["Sales"]).reset_index(drop=True)
```

```python
df
```

```python
df.info()
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"], format="%Y-%M-%d")
```

```python
df[["Segment", "Ship Mode"]] = df["Segment and Ship Mode"].str.split(" ", n=1, expand=True)
```

```python
df["Segment"] = df["Segment"].str.replace("_", " ")
```

```python
df = df.drop(columns={"Segment and Ship Mode"})
```
