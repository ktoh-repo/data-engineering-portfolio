```python
import pandas as pd
```

```python
df = pd.read_excel("../2.-Badly-Structured-Sales-Data-2.xlsx", header = 1)
```

```python
df
```

```python
df.columns
```

```python
df.rename(columns={"Segment": "Order Date", "Consumer": "First Class Consumer", "Corporate": "First Class Corporate", "Home Office": "First Class Home_Office",
                  "Consumer.1": "Same Day Consumer", "Corporate.1": "Same Day Corporate", "Home Office.1": "Same Day Home_Office",
                   "Consumer.2": "Second Class Consumer", "Corporate.2": "Second Class Corporate", "Home Office.2": "Second Class Home_Office",
                   "Consumer.3": "Standard Class Consumer", "Corporate.3": "Standard Class Corporate", "Home Office.3": "Standard Class Home_Office",
                  }, inplace = True)
```

```python
df
```

```python
test = df.melt(id_vars=["Order Date"], value_vars=["First Class Consumer","First Class Corporate","First Class Home_Office",
                                            "Same Day Consumer","Same Day Corporate","Same Day Home_Office",
                                            "Second Class Consumer","Second Class Corporate","Second Class Home_Office",
                                            "Standard Class Consumer","Standard Class Corporate","Standard Class Home_Office"
                                           ], var_name="Ship Mode and Segment", value_name="Sales")
```

```python
test[test["Sales"].isna() == False]
```

```python
test = test.dropna(subset=["Sales"]).reset_index(drop=True)
```

```python
test['Sales'] = round(test['Sales'],2)
```

```python
test[["Ship Mode", "Segment"]] = test["Ship Mode and Segment"].str.rsplit(" ", n=1, expand=True)
```

```python
test
```

```python
test['Segment'] = test['Segment'].str.replace("_", " ")
```

```python
test.drop(columns=["Ship Mode and Segment"], inplace=True)
```

```python
test.info()
```

```python
test['Order Date'] = pd.to_datetime(test['Order Date'])
```
