```python
import pandas as pd
```

```python
df = pd.read_csv("6M-0K-99K.users.dataset.public.csv")

df.head()
```

```python
df.count()
```

```python
import os 

os.makedirs("chunk-data")
os.listdir()
```

```python
# calculate the size of each chunk
chunk_size = len(df) // 5 + (1 if len(df) % 5 else 0)

# Split the dataframe into chunks and save them
for i in range(5):
    start = i * chunk_size
    end = start + chunk_size
    chunk_df = df.iloc[start:end]
    chunk_df.to_csv(f'chunk-data/chunk{i+1}.csv', index=False)
```
